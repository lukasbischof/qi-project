```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
title = 'Accuracy of Hybrid NN on AdHoc Datasets'
accuracies = {
    'AdHoc 100': [0.56, 0.52, 0.68, 0.48, 0.72, 0.56, 0.56, 0.52, 0.68, 0.56],
    'AdHoc 1k': [0.6200, 0.5600, 0.6200, 0.5920, 0.5760, 0.5040, 0.6000, 0.7200, 0.6520, 0.6240],
    'AdHoc 10k': [0.6204, 0.5348, 0.5764, 0.5936, 0.5064, 0.5276, 0.5516, 0.6028, 0.5232, 0.5904]
}
```

```python
minima = np.array([np.min(accuracies[key]) for key in accuracies]).min() - 0.1
maxima = np.array([np.max(accuracies[key]) for key in accuracies]).max() + 0.1

fig, axs = plt.subplots(1, len(accuracies), figsize=(3 * 2, 2 * 2))
fig.tight_layout(pad=2.0)
fig.suptitle(title)
fig.subplots_adjust(top=0.85)
for i, key in enumerate(accuracies):
    ax = axs[i]
    ax.boxplot(accuracies[key])
    ax.title.set_text(key)
    ax.set_xticklabels([])
    ax.set_ylim(minima, maxima)
    ax.grid()
plt.show()
```
