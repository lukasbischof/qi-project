```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
title = 'Accuracy of QNN on Custom Datasets'
accuracies = {
    'Custom 100': [0.72, 0.72, 0.72, 0.72, 0.72, 0.72, 0.72, 0.72, 0.72, 0.72],
    'Custom 1k': [0.728, 0.728, 0.728, 0.728, 0.728, 0.728, 0.728, 0.728, 0.728, 0.728],
    'Custom 10k': [0.7032, 0.7032, 0.7032, 0.7032, 0.7032, 0.7032, 0.7032, 0.7032, 0.7032, 0.7032]
}
```

```python
minima = np.array([np.min(accuracies[key]) for key in accuracies]).min() - 0.1
maxima = np.array([np.max(accuracies[key]) for key in accuracies]).max() + 0.1

fig, axs = plt.subplots(1, len(accuracies), figsize=(3 * 2, 2 * 2))
fig.tight_layout(pad=2.0)
fig.suptitle(title)
fig.subplots_adjust(top=0.85)
for i, key in enumerate(accuracies):
    ax = axs[i]
    ax.boxplot(accuracies[key])
    ax.title.set_text(key)
    ax.set_xticklabels([])
    ax.set_ylim(minima, maxima)
    ax.grid()
plt.show()
```

```python
print(f"Mean 'Custom 100': {statistics.mean(accuracies['Custom 100'])}")
print(f"Std Dev. 'Custom 100': {statistics.stdev(accuracies['Custom 100'])}")
print(f"Mean 'Custom 1k': {statistics.mean(accuracies['Custom 1k'])}")
print(f"Std Dev. 'Custom 1k': {statistics.stdev(accuracies['Custom 1k'])}")
print(f"Mean 'Custom 10k': {statistics.mean(accuracies['Custom 10k'])}")
print(f"Std Dev. 'Custom 10k': {statistics.stdev(accuracies['Custom 10k'])}")
```
